```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = -1
u0 = 1
t_max = 30
t = np.arange(0, t_max, 0.1)
dt1 = 1.1
dt2 = 1.5
dt3 = 1.9
```

```python
sol = np.exp(a*t)
```

```python
u_k1 = [u0]
u_k2 = [u0]
u_k3 = [u0]

for i in range(int(t_max/dt1)):
    u_1 = u_k1[-1]
    u_k1.append(u_1 * (1 + a*dt1))

for i in range(int(t_max/dt2)):
    u_2 = u_k2[-1]
    u_k2.append(u_2 * (1 + a*dt2))

for i in range(int(t_max/dt3)):
    u_3 = u_k3[-1]
    u_k3.append(u_3 * (1 + a*dt3))
```

```python
plt.plot(t, sol, label="Solucin exacta")
plt.plot(dt1 * np.arange(len(u_k1)), u_k1, label=f"dt={dt1}")
plt.plot(dt2 * np.arange(len(u_k2)), u_k2, label=f"dt={dt2}")
plt.plot(dt3 * np.arange(len(u_k3)), u_k3, label=f"dt={dt3}")
plt.xlabel("t(s)")
plt.ylabel("u(t)")
plt.legend()
plt.show()
```
