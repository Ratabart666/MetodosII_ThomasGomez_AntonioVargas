```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
def f(x,y):
    return x + 2*y
```

```python
def Solucion(x):
    return -0.5*x - (1-np.exp(2*x))/4.
```

```python
N = 10
t = np.linspace(0,1,N)
```

```python
h = t[1] - t[0]
```

```python
yext = Solucion(t)
```

```python
plt.scatter(t,yext)
```

```python
def GetEuler(x,y,h):
    
    for i in range(1,len(x)):
        
        k1 = f(x[i-1],y[i-1])
        y[i] = y[i-1] + h*k1
```

```python
y1 = np.zeros_like(t)
y1[0] = 0
```

```python
GetEuler(t,y1,h)
```

```python
plt.plot(t,yext)
plt.plot(t,y1)
```

```python
def GetEuler2(x,y,h):
    
    for i in range(1,len(x)):
        
        k1 = f(x[i-1],y[i-1])
        k2 = f(x[i-1]+h,y[i-1] + h*k1)
        
        y[i] = y[i-1] + 0.5*h*(k1+k2)
```

```python
def f(x,y):
    return x + 2*y
```

```python
y2 = np.zeros_like(t)
y2[0] = 0.
```

```python
GetEuler2(t,y2,h)
```

```python
plt.plot(t,yext)
plt.plot(t,y1,label='Euler1')
plt.plot(t,y2,label='Euler2')
plt.legend()
```

# Runge Kutta O3

```python
def Runge_Kutta3(x,y,h):
    
    for i in range(1,len(x)):
        
        k1 = f(x[i-1],y[i-1])
        k2 = f(x[i-1]+(h/2),y[i-1] + (h/2)*k1)
        k3 = f(x[i-1]+h,y[i-1]-h*k1+2*h*k2)
        
        y[i] = y[i-1] + (h/6)*(k1+4*k2+k3)
```

```python
y3 = np.zeros_like(t)
y3[0] = 0.
```

```python
Runge_Kutta3(t,y3,h)
```

```python
plt.plot(t,yext)
plt.plot(t,y1,label='Euler1')
plt.plot(t,y2,label='Euler2')
plt.plot(t,y3,label='Euler3')
plt.legend()
```

# Orden 4

```python
def Runge_Kutta4(x,y,h):

    for i in range(1,len(x)):
        k1 = f(x[i-1], y[i-1])
        k2 = f(x[i-1]+(h/2),y[i-1] + (h/2)*k1)
        k3 = f(x[i-1]+(h/2),y[i-1] + (h/2)*k2)
        k4 = f(x[i-1] + h, y[i-1] + k3*h)
        
        y[i] = y[i-1] + (h/6)*(k1 + 2*k2 + 2*k3 + k4)
```

```python
y4 = np.zeros_like(t)
y4[0] = 0.
```

```python
Runge_Kutta4(t,y4,h)
```

```python
plt.plot(t,yext)
plt.plot(t,y1,label='Euler1')
plt.plot(t,y2,label='Euler2')
plt.plot(t,y3,label='Euler3')
plt.plot(t,y4,label='Euler4')
plt.legend()
```
