```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(u,t,q):
    return u**q
```

```python
u = np.linspace(0, 100, 200)
t0 = np.zeros_like(u)
t1 = np.zeros_like(u)
t2 = np.zeros_like(u)
t3 = np.zeros_like(u)
t4 = np.zeros_like(u)
t5 = np.zeros_like(u)
h = u[1] - u[0]
q = [0., 0.2, 0.4, 0.7, 0.9, 1.]
```

```python
def GetEuler2(u,t,h,q):
    t[0] = 1
    for i in range(1,len(u)):
        
        k1 = f(u[i-1],t[i-1], q)
        k2 = f(u[i-1]+h,t[i-1] + h*k1, q)
        
        t[i] = t[i-1] + 0.5*h*(k1+k2)
```

```python
GetEuler2(u,t0,h,q[0])
GetEuler2(u,t1,h,q[1])
GetEuler2(u,t2,h,q[2])
GetEuler2(u,t3,h,q[3])
GetEuler2(u,t4,h,q[4])
GetEuler2(u,t5,h,q[5])
```

```python
def sol(u):
    return np.e**u
sol = sol(u)
```

```python
plt.plot(u,t0, label="q=0.")
plt.plot(u,t1, label="q=0.2")
plt.plot(u,t2, label="q=0.4")
plt.plot(u,t3, label="q=0.7")
plt.plot(u,t4, label="q=0.9")
plt.plot(u,t5, label="q=1.")
plt.plot(u,sol, label="Solución 1")
plt.xlim(0,20)
plt.ylim(0,20)
plt.xlabel('t(s)')
plt.ylabel('u(t)')
plt.legend()
plt.show()
```
