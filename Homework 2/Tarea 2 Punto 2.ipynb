```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("ggplot")
```

```python
x0 = 1.0
v0 = 1.0
omega = np.pi
h = 0.01
```

```python
t = np.arange(0, 10, h)
x = np.zeros_like(t)
v = np.zeros_like(t)
```

```python
def M_Verlet(x, v, t, omega, h):
    
    x_ = x + v*h + (-omega**2*x)*h**2
    v_ = v + ((-omega**2*x))*h
    
    return x_, v_

x[0] = x0
v[0] = v0 

for i in range(1,len(t)):
    x[i], v[i] = M_Verlet(x[i-1], v[i-1], t[i-1], omega, h)
```

```python
def No_Verlet(x,t):
    return x*np.cos(np.pi*t)

x_nv = No_Verlet(x0,t)
```

```python
plt.plot(t, x,'g', label='Verlet')
plt.plot(t, x_nv,'r', label='Solution')
plt.xlabel('t(s)')
plt.ylabel('x(m)')
plt.legend()
plt.show()
```

EL mtodo es altamente estable ya que oscila alrededor de la solucin anlitica. Como podemos ver el algoritmo tambien es consistente.
