```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("ggplot")
```

```python
def f(x,y):
    return (x*y**2) - (2*y/x) - (1/x**3)
```

```python
x = np.linspace(np.sqrt(2), 100, 500)
y = np.zeros_like(x)
h = x[1] - x[0]
```

```python
def GetEuler2(x,y,h):
    
    for i in range(1,len(x)):
        
        k1 = f(x[i-1],y[i-1])
        k2 = f(x[i-1]+h,y[i-1] + h*k1)
        
        y[i] = y[i-1] + 0.5*h*(k1+k2)
```

```python
GetEuler2(x,y,h)
```

```python
def sol(x):
    return (2 - x**2)/(2*x**2 + x**4)

sol = sol(x)
```

```python
plt.scatter(x, y,color='r', label='Runge Kutta 2')
plt.plot(x, sol,color='b', label='Analytic solution')
plt.xlim(0,10)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

 Note que el método de rungekutta 2 es altamente eficiente computacionalmente, ya que necesita pocos puntos ( igualmente equisespaciados) para acercarse demasiado bien a la solución. Si comparamos con el método de euler, vemos que este último es muy ineficiente, requiriendo muchos mas puntos equisespaciados para acercarse a la solución
