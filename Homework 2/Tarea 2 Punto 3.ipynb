```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x,y):
    return (x*y**2) - (2*y/x) - (1/x**3)
```

```python
x = np.linspace(np.sqrt(2), 100, 200)
y = np.zeros_like(x)
h = x[1] - x[0]
```

```python
def GetEuler2(x,y,h):
    
    for i in range(1,len(x)):
        
        k1 = f(x[i-1],y[i-1])
        k2 = f(x[i-1]+h,y[i-1] + h*k1)
        
        y[i] = y[i-1] + 0.5*h*(k1+k2)
```

```python
GetEuler2(x,y,h)
```

```python
def sol(x):
    return (2 - x**2)/(2*x**2 + x**4)

sol = sol(x)
```

```python
plt.plot(x, y,'r', label='Runge Kutta')
plt.plot(x, sol,'b', label='No RK')
plt.xlim(0,10)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show
```
